```python
#9. create a dataframe by combining 2 dictionaries.... using   pd.concat([df1, df2])
```

```python
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
```

```python
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}
```

```python
G= pd.DataFrame.from_dict(GermanCars)
G
```

```python
J= pd.DataFrame(japaneseCars)
J
```

```python
# concatenate 2 dataframes, G and J

# creating keys with   keys=['Germany', 'Japan']

GJ= pd.concat([G,  J],  keys=['Germany','Japan'])
GJ
```

```python
#10. Merge two data frames using the 2 dictionaries

Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
```

```python
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Horsepower': [141, 80, 182 , 160]}
```

```python
# in order to concatenate 2 dictionaries, you need to first convert them to dataframe

Cars1= pd.DataFrame(Car_Price)
Cars1
```

```python
Cars2= pd.DataFrame(car_Horsepower)
Cars2
```

```python
# you can only concatenate 2 dataframes, not 2 dictionaries

Cars= pd.concat([Cars1, Cars2.iloc[:,1]], axis=1)
Cars
```

```python
# alternate way to merge two dataframes is...

Carss= pd.merge(Cars1, Cars2, on="Company")
Carss
```
